```python
import os

traing_image = os.listdir('/home/team116/lab/LAB_0/resnet/data/training_image')
validation_image = os.listdir('/home/team116/lab/LAB_0/resnet/data/validation_image')
print(traing_image)
print(validation_image)
```

```python
#.jpg 
traing_image = [x.split(".")[0] for x in traing_image]
#print(traing_image)
#traing_image  _ split
training_image = [x.split("_") for x in traing_image]
print(training_image)
```

```python
#.jpg 
validation_image = [x.split(".")[0] for x in validation_image]
#print(validation_image)
#traing_image  _ split
validation_image = [x.split("_") for x in validation_image]
print(validation_image)
```

```python
#pandas dataframe 
import pandas as pd

training_image = pd.DataFrame(training_image, columns = ["W/T", "imageID", "Era", "style", "W/M"])
traing_image
#print(type(training_image))
```

```python
train_grouped_df = training_image.groupby(['W/M', 'style']).size().reset_index(name='count')
print(train_grouped_df)
```

```python
validation_image = pd.DataFrame(validation_image, columns = ["W/T", "imageID", "Era", "style", "W/M"])
validation_image
```

```python
validation_grouped_df = validation_image.groupby(['W/M', 'style']).size().reset_index(name='count')
print(validation_grouped_df)
```
