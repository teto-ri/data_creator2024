```python
#get filename list from data folder

import os
import sys

def get_file_list(path):
    file_list = []
    for root, dirs, files in os.walk(path):
        for file in files:
            file_list.append(file)
    return file_list

training_image_list = get_file_list('data/training_image')

validation_image_list = get_file_list('data/validation_image')

#{W/T}_{이미지ID}_{시대별}_{스타일별}_{성별}.jpg in list to dataframe
def get_image_info(file_list):
    image_info = []
    for file in file_list:
        #remove .jpg filter
        file = file.replace('.jpg', '')
        image_info.append(file.split('_'))
    return image_info

training_image_info = get_image_info(training_image_list)
validation_image_info = get_image_info(validation_image_list)

#to dataframe
import pandas as pd

training_df = pd.DataFrame(training_image_info, columns=['WT', 'ID', '시대', '스타일', '성별'])
validation_df = pd.DataFrame(validation_image_info, columns=['WT', 'ID', '시대', '스타일', '성별'])

display(training_df, validation_df)

# group by 성별, 스타일 count
training_grouped = training_df.groupby('성별')['스타일'].value_counts().to_frame(name='이미지 수')
validation_grouped = validation_df.groupby('성별')['스타일'].value_counts().to_frame(name='이미지 수')

display(training_grouped, validation_grouped)
```
