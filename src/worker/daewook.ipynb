2-1. 주어진 라벨링 데이터의 파일 명은 아래와 같은 형식이다 . “ {W/T}_ { 이미지 ID}_{ 시대별 }_{ 스타일별 }_{ 성별 }_{ 설문 ID}.json ”이에 기반 하여 “설문 ID ” 수 기준으로 “성별 & 스타일” 통계치를 아래 표 형식 으로 기입한다 .

```python
training_label = os.listdir('./data/training_label')
validation_label = os.listdir('./data/validation_label')
```

```python
column_name_2 = ['W/T', 'image_id', 'period', 'style', 'Sex', 'survey_id']

training_label_data = {'file_name': training_label}
training_label_df = pd.DataFrame(training_label_data)
training_label_df[column_name_2] = training_label_df['file_name'].str.replace('.json', '').str.split('_', expand=True)

validation_label_data = {'file_name': validation_label}
validation_label_df = pd.DataFrame(validation_label_data)
validation_label_df[column_name_2] = validation_label_df['file_name'].str.replace('.json', '').str.split('_', expand=True)

display(training_label_df, validation_label_df)
```

```python
drop_training_label_df = training_label_df[training_label_df['image_id'].isin(training_df['image_id'])]
drop_validation_label_df = validation_label_df[validation_label_df['image_id'].isin(validation_df['image_id'])]

display(drop_training_label_df, drop_validation_label_df)
```

```python
drop_training_label_df.loc[:, 'T/V'] = 'Train'
drop_validation_label_df.loc[:, 'T/V'] = 'Validation'
```

```python
import pandas as pd
import json

train_path = './data/training_label'
valid_path = './data/validation_label'

# JSON 파일을 읽는 함수를 정의
def extract_json_data(file_name):
    with open(file_name, 'r') as f:
        data = json.load(f)
    
    # 필요한 데이터 추출
    R_id = data['user']['R_id']
    Q5 = data['item']['survey']['Q5']
    
    return R_id, Q5

# drop_training_label_df에 R_id와 Q5 값을 담을 새로운 열을 생성
drop_training_label_df.loc[:, 'R_id'] = None
drop_training_label_df.loc[:, 'Q5'] = None
drop_validation_label_df.loc[:, 'R_id'] = None
drop_validation_label_df.loc[:, 'Q5'] = None

# file_name을 바탕으로 JSON 파일에서 데이터를 추출하고 DataFrame에 추가
for index, row in drop_training_label_df.iterrows():

    file_name = os.path.join(train_path, row['file_name'])
    R_id, Q5 = extract_json_data(file_name)
    
    # DataFrame에 값 추가
    drop_training_label_df.at[index, 'R_id'] = R_id
    drop_training_label_df.at[index, 'Q5'] = Q5


for index, row in drop_validation_label_df.iterrows():

    file_name = os.path.join(valid_path, row['file_name'])
    R_id, Q5 = extract_json_data(file_name)
    
    # DataFrame에 값 추가
    drop_validation_label_df.at[index, 'R_id'] = R_id
    drop_validation_label_df.at[index, 'Q5'] = Q5

display(drop_training_label_df, drop_validation_label_df)
```

■ 3-1. 추천 시스템의 기본인 협업 필터링 (Collaborative Filtering) 은 크게 user-based filtering, item-based filtering 방식으로 나뉘어져 있다 . 각각에 대해서 이해하고 , 2-2 에서 구해 본 응답자의 “스타일 선호 정보표”를 토대로 Validation 데이터 내 응답자의 “스타일 선호 여부 예측” 문제를 2 가지 기법으로 어떻게 적용해 볼 수 있고 , 서로 비교하여 어떤 장단점을 갖는지 설명한다 .

※ 설명을 용이하게 하기 위해 응답자의 스타일 선호도 예시를 들어서 설명해도 무방하다.

협업 필터링이란 사용자와 아이템 간의 상호작용 데이터를 기반으로 유사한 사용자나 아이템을 찾아 새로운 추천을 생성하는 방법으로 사용자의 과거 행동 데이터나 유사한 사용자 그룹의 데이터를 활용해 개인화된 추천을 제공하는 것을 말합니다.

협업 필터링은 "User-based Filtering", "Item-based Filtering" 로 나누어져있습니다. 

1. User-based Filtering
+ User-based filtering은 사용자의 취향이 비슷한 다른 사용자의 선호도를 바탕으로 추천을 진행합니다. 
+ 즉, 응답자 간의 유사성을 계산하고, 유사한 응답자가 선호하는 스타일을 현재 응답자에게 추천하거나 선호 여부를 예측하는 방식입니다.

적용 방법
+ 유사도 계산: 각 응답자에 대해 선호 스타일 정보를 벡터화한 뒤, 다른 응답자와의 유사도를 계산합니다. 일반적으로 코사인 유사도(Cosine Similarity) 또는 피어슨 상관계수(Pearson Correlation)를 사용하여 유사도를 구할 수 있습니다.
+ 유사한 응답자 찾기: 계산된 유사도에 따라 현재 응답자와 가장 유사한 응답자 그룹을 찾습니다.
+ 스타일 선호 예측: 유사한 응답자가 선호하는 스타일을 기반으로 Validation 데이터 내 응답자의 스타일 선호 여부를 예측합니다. 예를 들어, 유사한 응답자가 선호하는 스타일이 Validation 데이터에 포함되어 있다면 이를 선호할 확률이 높다고 예측합니다.

장단점
+ 장점:
    + 다양성: 개인화된 추천을 제공하기 때문에, 유사한 응답자들이 선호하는 스타일을 쉽게 예측할 수 있습니다.
    + 새로운 아이템 예측 가능: Validation 데이터에 포함된 새로운 스타일이라도 유사한 응답자의 선호를 토대로 예측이 가능합니다.

+ 단점:
    + 계산 비용: 사용자가 많아질수록 유사도 계산 비용이 증가합니다. 특히, 데이터가 큰 경우 효율성이 떨어집니다.
    + 희소성 문제: 사용자가 선호하는 스타일이 적으면, 유사도 계산이 어렵거나 잘못된 결과가 나올 수 있습니다.

2. Item-based Filtering
+ Item-based filtering은 스타일 간의 유사도를 계산하여 추천을 진행하는 방식입니다. 
+ 즉, 응답자가 특정 스타일을 선호한다면, 해당 스타일과 유사한 스타일을 선호할 가능성이 높다고 예측하는 방식입니다.

적용 방법
+ 유사도 계산: 각 스타일을 벡터화하여 스타일 간의 유사도를 계산합니다. 여기에서도 코사인 유사도 또는 피어슨 상관계수를 사용할 수 있습니다.
+ 유사한 스타일 찾기: 응답자가 선호하는 스타일과 유사한 스타일을 찾아 Validation 데이터에서 해당 스타일에 대한 선호 여부를 예측합니다.
+ 스타일 선호 예측: 유사한 스타일을 선호할 가능성이 높다고 예측하여, Validation 데이터에서 선호 여부를 추정합니다.

장단점
+ 장점:
    + 응답자 수가 많아도 안정적: 스타일 자체를 기반으로 추천을 진행하기 때문에, 응답자 수가 증가해도 계산 효율이 높습니다.
    + 스타일 예측의 정확성: 스타일 간의 유사성을 기반으로 예측하기 때문에, 스타일에 대한 취향이 분명한 경우 정확도가 높아집니다.

+ 단점:
    + 새로운 스타일 처리 불가: Validation 데이터에 포함된 새로운 스타일에 대해 유사도를 계산할 수 없기 때문에, 예측이 어렵습니다.
    + 응답자 취향 반영 어려움: 개별 응답자의 특성을 반영하기 어려워, 개인화된 추천에는 한계가 있습니다.

예시를 통한 설명
응답자 A가 힙합 스타일과 캐주얼 스타일을 선호한다고 가정해보겠습니다.

+ User-based Filtering을 적용할 경우: 유사한 선호도를 보이는 응답자 B, C가 힙합과 캐주얼 스타일 외에 스포티 스타일을 선호하는 것을 발견했다면, A도 스포티 스타일을 선호할 가능성이 있다고 예측합니다.
+ Item-based Filtering을 적용할 경우: A가 선호하는 힙합 스타일과 유사한 스트릿 패션 스타일이 Validation 데이터에 포함되어 있다면, A가 스트릿 패션 스타일도 선호할 가능성이 있다고 예측합니다.

결론
+ User-based Filtering은 개별 응답자의 선호도를 더 잘 반영할 수 있어 개인화에 유리하지만, 데이터가 희소할 때 예측 성능이 떨어질 수 있습니다.
+ Item-based Filtering은 응답자 수가 증가해도 효율성이 유지되며 스타일 예측이 명확해지지만, 개인화된 추천에는 한계가 있을 수 있습니다.
