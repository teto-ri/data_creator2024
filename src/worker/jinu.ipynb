# Mission 2-2

```python
# new_train_label 'dataset'   'train' 
new_training_label = new_training_label.assign(dataset='train')

# new_validation_label 'dataset'   'valid' 
new_validation_label = new_validation_label.assign(dataset='valid')

#  DataFrame concat 
label = pd.concat([new_training_label, new_validation_label], ignore_index=True)

#  
print(label)
```

```python
import pandas as pd
import os
import json

#    (  )
training_directory = '/home/team116/lab/LAB_0/resnet/data/training_label'  # Training   
validation_directory = '/home/team116/lab/LAB_0/resnet/data/validation_label'  # Validation   

# label  r_id Q5    
r_ids = []
q5_values = []


for index, row in label.iterrows():
    file_name = row['fileName']  
    dataset = row['dataset']  

    #   
    if dataset == 'train':
        file_path = os.path.join(training_directory, file_name)
    elif dataset == 'valid':
        file_path = os.path.join(validation_directory, file_name)
    
    # JSON  
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            
            # R_id 
            r_id = data['user']['R_id']
            r_ids.append(r_id)  # r_id  

            # Q5  
            q5 = data['item']['survey']['Q5']
            q5_values.append(q5)  # Q5  
    
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        r_ids.append(None)  #    None 
        q5_values.append(None)  #    None 
    except json.JSONDecodeError:
        print(f"Error decoding JSON from file: {file_path}")
        r_ids.append(None)  # JSON     None 
        q5_values.append(None)  # JSON     None 

# label  r_id Q5  
label['R_id'] = r_ids
label['Q5'] = q5_values

label
```

```python
# imageName  
label['imageName'] = label['W/T'] + "_" + label['imageID'].astype(str) + "_" + label['Era'].astype(str) + "_" + label['style'] + "_" + label['W/M'] + ".jpg"

label
```

```python
#R_id unique  
label['R_id'].nunique()
```

```python
label.info()
```

```python
training_data = label[label['dataset'] == 'train']
validation_data = label[label['dataset'] == 'valid']

#print(training_data['R_id'].nunique())
#print(validation_data['R_id'].nunique())

result =pd.DataFrame ({
    "Training  " : training_data[training_data['Q5'] == 2].groupby('R_id')['imageName'].apply(list),
    "Training  " : training_data[training_data['Q5'] == 1].groupby('R_id')['imageName'].apply(list),
    "Validation  " : validation_data[validation_data['Q5'] == 2].groupby('R_id')['imageName'].apply(list),
    "Validation  " : validation_data[validation_data['Q5'] == 1].groupby('R_id')['imageName'].apply(list)
})

result
```
