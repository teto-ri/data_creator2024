```python
import os
import json
import pandas as pd

# 1. validation_label 디렉토리 내의 모든 .json 파일에서 E_id, R_id, Q5를 추출
json_dir = "/home/team116/lab/LAB_0/resnet/data/validation_label"  # json 파일이 저장된 디렉토리
e_id_r_id_data = []

# 2. 각 json 파일에서 E_id, R_id, Q5를 추출
for json_file in os.listdir(json_dir):
    if json_file.endswith('.json'):
        with open(os.path.join(json_dir, json_file), 'r') as f:
            json_data = json.load(f)
            E_id = str(json_data.get('E_id'))  # E_id를 문자열로 변환
            R_id = json_data['user'].get('R_id')
            Q5 = json_data['item']['survey'].get('Q5')  # Q5 값 추출

            # E_id가 6자리가 아닐 경우 0으로 채우기
            E_id = E_id.zfill(6)

            # 데이터 리스트에 추가
            e_id_r_id_data.append({'E_id': E_id, 'R_id': R_id, 'Q5': Q5})

# 3. e_id_r_id_data를 DataFrame으로 변환
e_id_r_id_df = pd.DataFrame(e_id_r_id_data)

# 5. e_id_r_id_df의 E_id도 문자열로 변환 (이미 문자열로 처리됨)
e_id_r_id_df['E_id'] = e_id_r_id_df['E_id'].astype(str).str.strip()

# 6. vali_label_df에서 기존의 R_id 열 삭제 (중복을 피하기 위해)
if 'R_id' in vali_label_df.columns:
    vali_label_df = vali_label_df.drop(columns=['R_id'])

# 7. surveyID와 E_id를 매칭하여 R_id와 Q5 추가 (left join)
vali_label_df = vali_label_df.merge(
    e_id_r_id_df, 
    how='left', 
    left_on='surveyID', 
    right_on='E_id'
)

# 8. 필요 없는 E_id 컬럼 삭제
vali_label_df = vali_label_df.drop(columns=['E_id'])

# 결과 확인
vali_label_df

#R_id를 기준으로 groupby


```

```python
json_dir = "/home/team116/lab/LAB_0/resnet/data/training_label"  # json 파일이 저장된 디렉토리
e_id_r_id_data = []

# 2. 각 json 파일에서 E_id, R_id, Q5를 추출
for json_file in os.listdir(json_dir):
    if json_file.endswith('.json'):
        with open(os.path.join(json_dir, json_file), 'r') as f:
            json_data = json.load(f)
            E_id = str(json_data.get('E_id'))  # E_id를 문자열로 변환
            R_id = json_data['user'].get('R_id')
            Q5 = json_data['item']['survey'].get('Q5')  # Q5 값 추출

            # E_id가 6자리가 아닐 경우 0으로 채우기
            E_id = E_id.zfill(6)

            # 데이터 리스트에 추가
            e_id_r_id_data.append({'E_id': E_id, 'R_id': R_id, 'Q5': Q5})

# 3. e_id_r_id_data를 DataFrame으로 변환
e_id_r_id_df = pd.DataFrame(e_id_r_id_data)

# 5. e_id_r_id_df의 E_id도 문자열로 변환 (이미 문자열로 처리됨)
e_id_r_id_df['E_id'] = e_id_r_id_df['E_id'].astype(str).str.strip()

# 6. vali_label_df에서 기존의 R_id 열 삭제 (중복을 피하기 위해)
if 'R_id' in train_label_df.columns:
    train_label_df = train_label_df.drop(columns=['R_id'])

# 7. surveyID와 E_id를 매칭하여 R_id와 Q5 추가 (left join)
train_label_df = train_label_df.merge(
    e_id_r_id_df, 
    how='left', 
    left_on='surveyID', 
    right_on='E_id'
)

# 8. 필요 없는 E_id 컬럼 삭제
train_label_df = train_label_df.drop(columns=['E_id'])

# 결과 확인
train_label_df
```

```python
import pandas as pd

# 1. Training과 Validation 데이터 나누기
train_data = train_label_df[train_label_df['train/vali'] == 'train']
vali_data = vali_label_df[vali_label_df['train/vali'] == 'vali']

# 2. Training 데이터에서 스타일 선호/비선호 나누기 (imageID만 함께 저장)
train_style_pref = train_data[train_data['Q5'] == 2].groupby('R_id').apply(
    lambda x: list(x['imageID'])
)  # 선호 (imageID만 저장)

train_style_nonpref = train_data[train_data['Q5'] == 1].groupby('R_id').apply(
    lambda x: list(x['imageID'])
)  # 비선호 (imageID만 저장)

# Validation 데이터에서 스타일 선호/비선호 나누기 (imageID만 함께 저장)
# 1. Q5 == 2 (선호)일 때만 그룹화
vali_style_pref = vali_data[vali_data['Q5'] == 2].groupby('R_id').apply(
    lambda x: list(x['imageID']) if len(x) > 0 else []
)

# 2. Q5 == 1 (비선호)일 때만 그룹화
vali_style_nonpref = vali_data[vali_data['Q5'] == 1].groupby('R_id').apply(
    lambda x: list(x['imageID']) if len(x) > 0 else []
)

# 3. R_id로 데이터 병합 (outer join을 사용하여 R_id가 다르더라도 포함)
final_table = pd.DataFrame({'R_id': list(set(train_style_pref.index) | set(train_style_nonpref.index) | set(vali_style_pref.index) | set(vali_style_nonpref.index))})

final_table = final_table.merge(train_style_pref.rename('Training Style 선호 (imageID)'), on='R_id', how='left')
final_table = final_table.merge(train_style_nonpref.rename('Training Style 비선호 (imageID)'), on='R_id', how='left')
final_table = final_table.merge(vali_style_pref.rename('Validation Style 선호 (imageID)'), on='R_id', how='left')
final_table = final_table.merge(vali_style_nonpref.rename('Validation Style 비선호 (imageID)'), on='R_id', how='left')

# 4. NaN을 빈 리스트로 변환 (필요한 경우)
final_table = final_table.fillna(' ')

# 결과 확인
final_table
```

```python
# imageID_list를 DataFrame으로 변환
imageID=[i[2:7] for i in imageID_list]
imageID_df = pd.DataFrame({'filename': imageID_list})
imageID_df['imageID']=imageID
imageID_df

imageID_to_filename = dict(zip(imageID_df['imageID'], imageID_df['filename']))

# final_table의 각 컬럼에 대해 imageID를 filename으로 변환
for column in ['Training Style 선호 (imageID)', 'Training Style 비선호 (imageID)', 
               'Validation Style 선호 (imageID)', 'Validation Style 비선호 (imageID)']:
    final_table[column] = final_table[column].apply(
        lambda x: [imageID_to_filename.get(imageID, imageID) for imageID in x] if isinstance(x, list) else x
    )

# 결과 확인
final_table
```
